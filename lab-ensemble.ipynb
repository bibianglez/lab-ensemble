```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
def df_info(df):
    print("\n    ---- df info ---")
    print(df.info())
    print("\n    ---- Empty Values ---")
    print(df.isna().sum())
    print("\n    ----- Dtypes -----")
    print(df.dtypes)
    print("\n    ---- Statistics ---")
    return df.describe()
df_info(spaceship)
```

```python
X=spaceship
```

```python
X.rename(columns={col : col.strip() for col in X.columns}, inplace = True)
X.rename(columns={col : col.replace(" ", "_").lower() for col in X.columns}, inplace = True)
X ["cabin"] = X ["cabin"].str[0]
X = X.dropna()
X.head()
```

```python
target = X["transported"]
features = X.drop(columns = ["passengerid","name", "transported"])
features = pd.get_dummies (features, drop_first = True)
```

**Perform Train Test Split**

```python
X_train,X_test, y_train,y_test = train_test_split(features,target,test_size=0.2,random_state=0)
X_train.head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

X_train_norm = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)

X_test_norm = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
display(X_train_norm.head(5))
display(X_test_norm.head(5))
```

**Model Selection** - now you will try to apply different ensemble methods in order to get a better model

- Bagging and Pasting

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import f1_score, precision_score, recall_score

PARAMS_BAGGING = {
    'n_estimators': 100,
    'max_samples': 1000,
    'random_state': 42
}

PARAMS_TREE = {
    'max_depth': 20,
    'random_state': 42
}

# --- TRAINING ---
base_tree = DecisionTreeClassifier(**PARAMS_TREE)
bagging_clf = BaggingClassifier(estimator=base_tree, **PARAMS_BAGGING)
bagging_clf.fit(X_train_norm, y_train)

# --- PREDICTIONS AND METRICS ---
pred = bagging_clf.predict(X_test_norm)

print("Accuracy", bagging_clf.score(X_test_norm, y_test))
print("F1 score", f1_score(y_test, pred))
print("Precision", precision_score(y_test, pred))
print("Recall", recall_score(y_test, pred))

```

```python
# Pasting: sampling without replacement
bagging_pasting = BaggingClassifier(estimator=base_tree, bootstrap=False, **PARAMS_BAGGING)
bagging_pasting.fit(X_train_norm, y_train)
pred_pasting = bagging_pasting.predict(X_test_norm)

print("Accuracy", bagging_pasting.score(X_test_norm, y_test))
print("F1 score", f1_score(y_test, pred_pasting))
```

- Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, precision_score, recall_score

forest = RandomForestClassifier()
forest.fit(X_train_norm, y_train)
pred = forest.predict(X_test_norm)

#print("MAE", mean_absolute_error(pred, y_test))
#print("RMSE", np.sqrt(mean_squared_error(pred, y_test)))
print("Accuracy", forest.score(X_test_norm, y_test))
print("F1 score", f1_score(y_test, pred))
print("Precision", precision_score(y_test, pred))
print("Recall", recall_score(y_test, pred))
```

- Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import f1_score, precision_score, recall_score

gb_classifier = GradientBoostingClassifier()
gb_classifier.fit(X_train_norm, y_train)
pred = gb_classifier.predict(X_test_norm)

print("Accuracy", gb_classifier.score(X_test_norm, y_test))
print("F1 score", f1_score(y_test, pred))
print("Precision", precision_score(y_test, pred))
print("Recall", recall_score(y_test, pred))
```

- Adaptive Boosting

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, precision_score, recall_score

adapt_boost = AdaBoostClassifier(DecisionTreeClassifier())
adapt_boost.fit(X_train_norm, y_train)
pred = adapt_boost.predict(X_test_norm)  

print("Accuracy", adapt_boost.score(X_test_norm, y_test))
print("F1 score", f1_score(y_test, pred))
print("Precision", precision_score(y_test, pred))
print("Recall", recall_score(y_test, pred))
```
